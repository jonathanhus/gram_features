## Put It All Together

Run this to get scores for all results files

```python
import pandas as pd
from pathlib import Path
import lingtypology.glottolog as glotto

def process_results(input_file: str) -> int:
    # Handling cases for 4 columns or 5 columns
    four_cols = ['ID', 'code', 'comment', 'source']
    five_cols = ['ID', 'code', 'comment', 'source', 'prompt_info']

    df_test = pd.read_json(input_file, orient='index').reset_index()
    if df_test.shape[1] == len(four_cols):
        df_test.columns = four_cols
    elif df_test.shape[1] == len(five_cols):
        df_test.columns = five_cols

    # df_test.columns = ['ID', 'code', 'comment', 'source']

    # Merge the 2 tables
    merged_results = pd.merge(df_values, df_test, on="ID")
    merged_results['Value'] = merged_results['Value'].astype(str)
    merged_results['code'] = merged_results['code'].astype(str)

    match_results = merged_results[merged_results['Value'] == merged_results['code']]
    # print("Number of rows: ", match_results.shape[0])
    return(match_results.shape[0])

def get_total_features(glottocode: str) -> int:
    df_values = pd.read_csv('resources/grambank/values.csv')
    lang_features = df_values[df_values['Language_ID']==glottocode]
    return lang_features.shape[0]
```

```python
df_values = pd.read_csv('resources/grambank/values.csv')

scores = dict()
# for input_path in list(Path('outputs').iterdir()):
for input_path in [p for p in Path('outputs').iterdir() if p.is_dir()]:
    print(input_path)
    scores[input_path.name] = dict()
    for file in Path(input_path).iterdir():
        if file.name.endswith('RAG.json'):
            # print(file.name)
            chunks = file.name.split("_")[2]
            # print(f"NAME: {chunks}")
            scores[input_path.name][f'rag_{chunks}'] = process_results(file)
        elif file.name.endswith('results.json'):
            scores[input_path.name]['baseline'] = process_results(file)
        elif file.name.endswith('Query.json'):
            scores[input_path.name]['expansion'] = process_results(file)
        
        # Add total features for the language to the dataframe
        # Yes, it writes it twice, but I wanted to use the iso code
        # for the language
        glot_id = glotto.get_glot_id_by_iso(file.name.split("_")[0])
    # print(get_total_features(glot_id))
    scores[input_path.name]['total'] = get_total_features(glot_id)

print(scores)

df_hus = pd.DataFrame.from_dict(scores, orient='index')
print(df_hus)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
# data = {
#     'southern_jinghpaw': {'baseline': 51, 'rag': 101, 'total': 195},
#     'minangkabau': {'baseline': 90, 'rag': 83, 'total': 174},
#     'iloko': {'baseline': 48, 'rag': 54, 'total': 103},
#     'mizo': {'baseline': 89, 'rag': 112, 'total': 195},
#     'kalamang': {'baseline': 54, 'rag': 95, 'total': 195},
#     'iloko-forreal': {'baseline': 49, 'rag': 55, 'total': 103}
# }

data = scores

# Extract categories and values
categories = list(data.keys())
baseline_values = [data[group]['baseline'] for group in categories]
# rag_values = [data[group]['rag_4'] for group in categories]
# Get the best RAG score for each language
rag_values = [
    max(value for key, value in data[group].items() if key.startswith("rag_")) 
    for group in categories
]
expansion_values = [data[group]['expansion'] for group in categories]
total_values = [data[group]['total'] for group in categories]

# Calculate "extra" values for the stacked part
baseline_extra = [total - baseline for total, baseline in zip(total_values, baseline_values)]
rag_extra = [total - rag for total, rag in zip(total_values, rag_values)]
expansion_extra = [total - expansion for total, expansion in zip(total_values, expansion_values)]

# Define x positions for bars
x = np.arange(len(categories))  # The label locations
bar_width = 0.2  # Width of each bar

# Create the plot
fig, ax = plt.subplots(figsize=(7, 6))

# Plot Baseline Stacked Bars
# ax.bar(x - bar_width / 3, baseline_values, bar_width, label='Baseline', color='skyblue', edgecolor='black')
ax.bar(x - bar_width, baseline_values, bar_width, label='Baseline', color='skyblue', edgecolor='black')
ax.bar(x - bar_width, baseline_extra, bar_width, bottom=baseline_values, label='_nolegend_', color='white', edgecolor='black')

# Plot RAG Stacked Bars
ax.bar(x, rag_values, bar_width, label='RAG', color='orange', edgecolor='black')
ax.bar(x, rag_extra, bar_width, bottom=rag_values, label='_nolegend_', color='peachpuff', edgecolor='black')

# Plot Query Expansion Stacked Bars
ax.bar(x + bar_width, expansion_values, bar_width, label='Expansion', color='red',edgecolor='black')
ax.bar(x + bar_width, expansion_extra, bar_width, bottom=expansion_values, label='_nolegend_', color='pink', edgecolor='black')

# Add labels, title, and legend
# ax.set_xlabel('Groups')
ax.set_ylabel('Number of Features')
ax.set_title('Grambank Features Correctly Predicted')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='center')  # Rotate category labels for clarity
ax.legend()

# Add grid for better readability
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

### Performance vs Num of References

```python
import matplotlib.pyplot as plt

# Extract relevant data for the graph
x_labels = ['baseline'] + sorted(set(key for subdict in scores.values() for key in subdict if key.startswith('rag_')), key=lambda k: int(k.split('_')[1]))
x_indices = [label for label in x_labels]  # Full labels (used for retrieving data)
x_display = ['0' if label == 'baseline' else label.split('_')[1] for label in x_labels]  # Stripped labels for the x-axis

# Plot each language's data
plt.figure(figsize=(10, 6))
for language, metrics in scores.items():
    y_values = [metrics.get(label, None) for label in x_indices]  # Get y-values for the x-axis points
    plt.plot(x_display, y_values, marker='o', label=language)

# Customize the plot
plt.title('Feature Identification Based on Number of Retrieved Entries', fontsize=16)
plt.xlabel('# of Retrieved Chunks', fontsize=14)
plt.ylabel('Correctly Predicted Features', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Languages', loc='upper left', fontsize=10)
plt.grid(alpha=0.3)

# Show the plot
plt.tight_layout()
plt.show()
```
